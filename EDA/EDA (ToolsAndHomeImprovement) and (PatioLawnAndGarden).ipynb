#### Performing EDA analysis on Style column

```python
# Dropping null values to perform analysis on the style column
df_style_eda=df_tools_and_home.dropna()
```

```python
df_style_eda['style']
```

```python
# Checking type of data in style column 
for i in df_style_eda['style'][:3]:
    print(type(i))
```

###### We see that the data in the style column is of dictionary format but is of type string. Therefore we must convert it to a dictionary type

```python
# Converting string to dictionary and storing them in a list
import ast
style_dict=[]
for i in df_style_eda['style']:
    b=ast.literal_eval(i)
    style_dict.append(b)
style_dict
```

###### Checking for different style parameters from our Dictionaries

```python
# Checking for Style parameters
features=[]
for i in style_dict:
    for j,k in i.items():
        features.append(j)
set(features)
```

###### We shall consider only the style parameters whuch occur frequently i.e. : style, format, color, size, wattage

```python
style_count,format_count,color_count,size_count,wattage_count=[],[],[],[],[]
```

```python
for i in style_dict:
    for j,k in i.items():
        if (j=='Color:')|(j=='Color Name:'):
            color_count.append(k)
        else:
            color_count.append(np.nan)
            
        if j=='Format:':
            format_count.append(k)
        else:
            format_count.append(np.nan)
            
        if (j=='Style:')|(j=='Style Name:'):
            style_count.append(k)
        else:
            style_count.append(np.nan)
            
        if (j=='Size:')|(j=='Size Name:'):
            size_count.append(k)
        else:
            size_count.append(np.nan)
            
        if j=='Wattage:':
            wattage_count.append(k)
        else:
            wattage_count.append(np.nan)
```

###### We have now extracted our Color, Format, Style, Size, Wattage information from our style column

```python
# Checking if the length of items in our list is same
len(color_count),len(format_count),len(style_count),len(size_count),len(wattage_count)
```

###### We now store the information in a dataframe and proceed with the EDA

```python
# Storing in a dataframe to conduct EDA
df_style=pd.DataFrame()
df_style['Color']=color_count
df_style['Format']=format_count
df_style['Style']=style_count
df_style['Size']=size_count
df_style['Wattage']=wattage_count
```

```python
df_style_columns=list(df_style.columns)
df_style_columns
```

###### Checking in the value_counts for each column in our style dataframe to identify any noise in the data

```python
for i in df_style_columns:
    print(i)
    print('Value Counts:')
    print(df_style[i].value_counts())
    print('*'*50)
```

###### We see that there is presence of noise data in our Color column due to the presence of numbers.
###### Also there is a necessity for cleaning the size column due to no particular annotation being followed. For example: 1-pack also being represented as: 1 pack,1,pack of 1,1-(pack)

```python
c=[]
for i in df_style['Color']:
    str1=str(i)
    if any(map(str.isdigit,str1))==True:
        c.append(np.nan)
    else:
        c.append(i)
```

```python
df_style['Color']=c
```

```python
c1=[]
for i in df_style['Color']:
    c1.append(i)
set(c1)
```

```python
df_style['Size'].replace(to_replace=[' 1 Pack',' 1-(Pack)',' Pack of 1',' 1',' 1-Pack'],value='1-Pack',inplace=True)
```

```python
w=[]
for i in df_style['Wattage']:
    w.append(i)
set(w)
```

```python
def remove(string):
    return string.replace(' ','')
```

```python
w3=[]
for i in df_style['Wattage']:
    if i!=np.nan:
        str_w=remove(str(i))
        w3.append(str_w)
    else:
        w3.append(i)
```

```python
df_style['Wattage']=w3
```

```python
df_style['Wattage']=df_style['Wattage'].str.replace('[wW]','')
```

```python
df_style['Wattage']=df_style['Wattage'].astype('float64')
```

### Business Question 29.a) Represent the Sales in the Patio Lawn and Garden category with respect to time. Also identify if there is any trend or seasonality in the data

```python
time_data_patio.head()
```

```python
plt.figure(figsize=(20,12))
plt.plot(time_data_patio)
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
decomposition=seasonal_decompose(time_data_patio,period=1)
d_trend=decomposition.trend
d_seasonal=decomposition.seasonal
d_residual=decomposition.resid

# plotting all the 4 graphs
fig,ax = plt.subplots(4,1,figsize=(20,12))

plt.subplot(411)
plt.plot(time_data_patio,label='MRF Stock Data - as it is')
plt.legend(loc='best')

plt.subplot(412)
plt.plot(d_trend,label='Trend')
plt.legend(loc='best')

plt.subplot(413)
plt.plot(d_seasonal,label='Seasonal')
plt.legend(loc='best')

plt.subplot(414)
plt.plot(d_residual,label='Residual')
plt.legend(loc='best')

plt.tight_layout()
```
