```python
df_meta['price']=df_meta['price'].fillna('0')
```

```python
df_meta['price'].isnull().sum()
```

```python
df_meta['price']=df_meta['price'].str.replace('$','')
```

```python
df_meta['price']=df_meta['price'].str.replace(',','')
```

```python
p=[]
for i in df_meta['price']:
    if len(i)<10:
        p.append(i)
    else:
        p.append('0')
p
```

```python
df_meta['price']=p
```

```python
df_meta.dtypes
```

```python
df_meta['price']=df_meta['price'].astype('float64')
```

```python
lst_cat=list(df_meta['main_cat'].unique())
lst_cat
```

```python
df_AH=df_meta[(df_meta['main_cat']=='Amazon Home')&(df_meta['price']!=0.00)]
print(df_AH['price'].mean())
df_IS=df_meta[(df_meta['main_cat']=='Industrial & Scientific')&(df_meta['price']!=0.00)]
print(df_IS['price'].mean())
df_TG=df_meta[(df_meta['main_cat']=='Toys & Games')&(df_meta['price']!=0.00)]
print(df_TG['price'].mean())
df_THI=df_meta[(df_meta['main_cat']=='Tools & Home Improvement')&(df_meta['price']!=0.00)]
print(df_THI['price'].mean())
df_SO=df_meta[(df_meta['main_cat']=='Sports & Outdoors')&(df_meta['price']!=0.00)]
print(df_SO['price'].mean())
df_AB=df_meta[(df_meta['main_cat']=='All Beauty')&(df_meta['price']!=0.00)]
print(df_AB['price'].mean())
df_CP=df_meta[(df_meta['main_cat']=='Camera & Photo')&(df_meta['price']!=0.00)]
print(df_CP['price'].mean())
df_ACS=df_meta[(df_meta['main_cat']=='Arts, Crafts & Sewing')&(df_meta['price']!=0.00)]
print(df_ACS['price'].mean())
```

```python
AH=df_AH['price'].mean()
IS=df_IS['price'].mean()
TG=df_TG['price'].mean()
THI=df_THI['price'].mean()
SO=df_SO['price'].mean()
AB=df_AB['price'].mean()
CP=df_CP['price'].mean()
ACS=df_ACS['price'].mean()
```

```python
df_AH2=df_meta[df_meta['main_cat']=='Amazon Home']
df_IS2=df_meta[df_meta['main_cat']=='Industrial & Scientific']
df_TG2=df_meta[df_meta['main_cat']=='Toys & Games']
df_THI2=df_meta[df_meta['main_cat']=='Tools & Home Improvement']
df_SO2=df_meta[df_meta['main_cat']=='Sports & Outdoors']
df_AB2=df_meta[df_meta['main_cat']=='All Beauty']
df_CP2=df_meta[df_meta['main_cat']=='Camera & Photo']
df_ACS2=df_meta[df_meta['main_cat']=='Arts, Crafts & Sewing']
```

```python
df_AH2['price'].replace(to_replace=0.00,value=AH,inplace=True)
df_IS2['price'].replace(to_replace=0.00,value=IS,inplace=True)
df_TG2['price'].replace(to_replace=0.00,value=TG,inplace=True)
df_THI2['price'].replace(to_replace=0.00,value=THI,inplace=True)
df_SO2['price'].replace(to_replace=0.00,value=SO,inplace=True)
df_AB2['price'].replace(to_replace=0.00,value=AB,inplace=True)
df_CP2['price'].replace(to_replace=0.00,value=CP,inplace=True)
df_ACS2['price'].replace(to_replace=0.00,value=ACS,inplace=True)
```

```python
df_meta_concat=pd.concat([df_AH2,df_IS2,df_TG2,df_THI2,df_SO2,df_AB2,df_CP2,df_ACS2],axis=0)
```
